Notebook to create the plot of Fig 2.1 in the thesis

```python
import networkx as nx
import matplotlib.pyplot  as plt
```

```python
G = nx.Graph([(0, 1), (1, 2), (7, 0), (6,4), (3,6), (1,6), (0, 8), (6,5), (3,4), (2,9)])
```

```python
color_map = ['r', 'r', 'y', 'y', 'y', 'y', 'y', 'y', 'y', 'y']
nx.draw(G, with_labels=True,  node_color = color_map)
plt.savefig("plots/plots_net/central_node.pdf")
```
